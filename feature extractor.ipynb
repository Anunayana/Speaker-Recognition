```python
import os
from os import listdir
from os.path import isfile, join
#librosa feature extraction
import librosa
import numpy as np
import pandas as pd
```

```python
# home='Enter the audio path'
home='F:\\DATA SET\\voice_dataset'
```

```python
# list of all folder present in the subdirectory
test_directory = home
Sub_Directories=[]
for child in os.listdir(test_directory):
    test_path = os.path.join(test_directory, child)
    if os.path.isdir(test_path):
        print (test_path)
        Sub_Directories.append(test_path)
        # Do stuff to the directory "test_path"
```

```python
audio_path_all=[]
for file in range(len(Sub_Directories)):
    mypath=Sub_Directories[file]
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]   
    for j in range(len(onlyfiles)):
        audio_path=Sub_Directories[file]+'\\'+onlyfiles[j]
        audio_path_all.append(audio_path)
```

```python
audio_path_all
```

```python
chroma_stft_list=[]
spec_cent_list=[]
spec_bw_list=[]
rolloff_list=[]
zcr_list=[]
mfcc_list=[]
prognosis_list=[]

for i in range(len(audio_path_all)):
    audio = audio_path_all[i] 
    # y , sr = librosa.load(audio, mono=True)
    y, sr = librosa.load(audio, mono = True,duration=5.0)
    #features
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)

    
    #variables
    chroma_stft = np.mean(chroma_stft)
    spec_cent = np.mean(spec_cent)
    spec_bw = np.mean(spec_bw)
    rolloff = np.mean(rolloff)
    zcr = np.mean(zcr)
    mfcc = np.mean(mfcc)
    prognosis=audio_path_all[i].split('\\')[-2]
    
    chroma_stft_list.append(chroma_stft)
    spec_cent_list.append(spec_cent)
    spec_bw_list.append(spec_bw)
    rolloff_list.append(rolloff)
    zcr_list.append(zcr)
    mfcc_list.append(mfcc)
    prognosis_list.append(prognosis)

    
    
    
value={'chroma_stft':chroma_stft_list,'spec_cent':spec_cent_list,'spec_bw':spec_bw_list,'rolloff':rolloff_list,'zcr':zcr_list,'mfcc':mfcc_list,'prognosis':prognosis_list}
feature=pd.DataFrame(value)
```
